# Delta time calculation

Converts ISO timestamp to datetime and performs delta time calculation. Generates new file wil calculated time

```python
import pandas as pd

df = pd.read_csv("joints.csv")

# Convert ISO timestamp to datetime
df["timestamp"] = pd.to_datetime(df["timestamp"])

# Calculate delta time in seconds
df["delta_time"] = df["timestamp"].diff().dt.total_seconds()

# First row will be NaN â†’ set it to 0
df["delta_time"] = df["delta_time"].fillna(0)

df.to_csv("with_delta_time.csv", index=False)
```
